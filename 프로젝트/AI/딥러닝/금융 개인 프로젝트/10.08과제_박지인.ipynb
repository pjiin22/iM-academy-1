```python
items = driver.find_elements(By.CSS_SELECTOR, "div.product_item__MDtDF")  # 상품 리스트 가져오기

for item in items:
    try:
        title = item.find_element(By.CSS_SELECTOR, "div.product_title__Mmw2K").text  # 상품명 가져오기
        price = item.find_element(By.CSS_SELECTOR, "div.product_price_area__eTg7I").text  # 가격 가져오기
        star = item.find_element(By.CSS_SELECTOR, "span.basicList_star__3NkBn").text if item.find_elements(By.CSS_SELECTOR, "span.basicList_star__3NkBn") else "No Rating"  # 별점 가져오기
        bag_type = item.find_element(By.CSS_SELECTOR, "span.product_grade__IzyU3").text if item.find_elements(By.CSS_SELECTOR, "span.basicList_depth__2QIie") else "No Category"  # 카테고리 가져오기

        print(f"Title: {title}, Price: {price}, Star: {star}, Bag Type: {bag_type}")
    except Exception as e:
        print(f"Error extracting item: {e}")  # 오류 처리
```

```python
# 수집된 데이터를 완전히 초기화하는 함수
def reset_data():
    global box  # 전역 변수 box에 접근
    box.clear()  # 수집된 데이터를 모두 제거

# 수집을 멈추고 데이터를 초기화하고 다시 수집을 시작
reset_data()  # 수집 데이터를 초기화

print("데이터가 초기화되었습니다. 현재 box 길이:", len(box))
```

```python
box = []  

for _ in range(15):
    body = driver.find_element(By.CSS_SELECTOR, "body")
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)

    square = driver.find_elements(By.CSS_SELECTOR, "div.product_item__MDtDF")

    for bag in square:
    
        title = bag.find_element(By.CSS_SELECTOR, "div.product_title__Mmw2K").text
        
        price_element = bag.find_element(By.CSS_SELECTOR, "span.price_num__S2p_v em").text
        price = price_element.replace(",", "")  

        try:
            star = bag.find_element(By.CSS_SELECTOR, "span.product_grade__IzyU3").text.split("\n")[1]
        except:
            continue

        category_elements = bag.find_elements(By.CSS_SELECTOR, "span.product_category__l4FWz.product_nohover__Z0Muw")
        if len(category_elements) >= 3:
            bag_type = driver.execute_script("return arguments[0].textContent;", category_elements[2]).strip()
        else:
            bag_type = "No Category"

        
        data = {"상품명": title, "가격": price, "평점": star, "종류": bag_type}
        box.append(data)

for bag in box:
    print(bag)
```

```python
from sklearn.preprocessing import MinMaxScaler

bag = pd.DataFrame(box)  

bag = bag.dropna()

scaler = MinMaxScaler()
bag[['가격', '평점']] = scaler.fit_transform(bag[['가격', '평점']])

print(bag.head())
```

```python
import numpy as np
import pandas as pd

def euclidean_distance(bag1, bag2):
    price1, price2 = int(bag1['가격']), int(bag2['가격'])
    star1, star2 = float(bag1['평점']), float(bag2['평점'])
    
    distance = np.sqrt((price1 - price2)**2 + (star1 - star2)**2)
    return distance

def recommend_similar_bags(target_bag, data, n=5):
    distances = []

    for _, bag in data.iterrows():
        if bag['상품명'] != target_bag['상품명']:
            distance = euclidean_distance(target_bag, bag)
            distances.append((bag['상품명'], distance, bag['가격'], bag['평점'], bag['종류']))
    
    distances = sorted(distances, key=lambda x: x[1])
    recommended_bags = distances[:n]
    
    return recommended_bags

df['가격'] = df['가격'].replace(",", "", regex=True).astype(int)
df['평점'] = df['평점'].astype(float)

target_bag = df.iloc[0]
recommended_bags = recommend_similar_bags(target_bag, df, n=5)

print("추천된 가방 목록:")
print("{:<30} {:<10} {:<10} {:<10}".format("상품명", "가격", "평점", "종류"))
print("-" * 60)
for bag in recommended_bags:
    print("{:<30} {:<10} {:<10} {:<10}".format(bag[0], bag[2], bag[3], bag[4]))
```
