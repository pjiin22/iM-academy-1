## 크롭 안한 버전

```python
measurements_df['avg_moisture'] = measurements_df[['수분_이마', '수분_오른쪽볼', '수분_왼쪽볼']].mean(axis=1)

def categorize_moisture(moisture_value):
    if moisture_value <= 20:
        return 'Very Dry'
    elif 21 <= moisture_value <= 40:
        return 'Dry'
    elif 41 <= moisture_value <= 60:
        return 'Normal'
    elif 61 <= moisture_value <= 80:
        return 'Hydrated'
    else:
        return 'Over-hydrated'

measurements_df['moisture_category'] = measurements_df['avg_moisture'].apply(categorize_moisture)
```

```python
# 이미지와 출력 폴더 경로
image_folder = r'.\image'  # 이미지가 있는 경로
output_folder = r'.\output'  # 분류된 이미지가 저장될 경로

# 등급별 폴더 생성
categories = ['Very Dry', 'Dry', 'Normal', 'Hydrated', 'Over-hydrated']
for category in categories:
    os.makedirs(os.path.join(output_folder, category), exist_ok=True)

# 각 subject_no에 대한 이미지를 분류
for index, row in measurements_df.iterrows():
    subject_no = row['subject_no']
    moisture_category = row['moisture_category']
    
    # subject_no에 해당하는 폴더 (예: '0003')
    subject_folder = os.path.join(image_folder, str(subject_no).zfill(4))
    
    # "F"로 끝나는 이미지 파일명만 처리
    image_filename = f"{str(subject_no).zfill(4)}_03_F.jpg"
    
    # 원본 이미지 파일 경로
    source_image_path = os.path.join(subject_folder, image_filename)
    
    # 디버깅용 경로 출력
    print(f"Checking path: {source_image_path}")
    
    # 분류될 이미지 파일 경로 (수분 등급에 따른 폴더에 저장)
    destination_image_path = os.path.join(output_folder, moisture_category, image_filename)

    # 이미지 파일이 존재할 경우에만 복사 수행
    if os.path.exists(source_image_path):
        print(f"Found: {source_image_path}")
        shutil.copy(source_image_path, destination_image_path)
    else:
        print(f"Image not found: {source_image_path}")
```

```python
import os
import pandas as pd
from PIL import Image
import json

# 이미지와 출력 폴더 경로
image_folder = r'.\image'  # 이미지가 있는 경로
output_folder = r'.\output2'  # 정면 이마 이미지가 저장될 경로
bbox_folder = r'.\bbox'  # bbox 정보가 있는 JSON 파일 경로

# 등급별 폴더 생성
categories = ['Very Dry', 'Dry', 'Normal', 'Hydrated', 'Over-hydrated']
for category in categories:
    os.makedirs(os.path.join(output_folder, category), exist_ok=True)

# 수분 값을 기준으로 분류 함수
def categorize_moisture(moisture_value):
    if moisture_value <= 20:
        return 'Very Dry'
    elif 21 <= moisture_value <= 40:
        return 'Dry'
    elif 41 <= moisture_value <= 60:
        return 'Normal'
    elif 61 <= moisture_value <= 80:
        return 'Hydrated'
    else:
        return 'Over-hydrated'

# 정면 이마(F) 처리 함수
def process_forehead(subject_no, moisture_value):
    json_filename = f"{str(subject_no).zfill(4)}_03_F_01.json"  # 정면 이마 JSON 파일명
    json_path = os.path.join(bbox_folder, str(subject_no).zfill(4), json_filename)

    # JSON 파일에서 bbox 정보 가져오기
    if os.path.exists(json_path):
        with open(json_path, 'r') as f:
            data = json.load(f)
            bbox = data['images']['bbox']
    else:
        print(f"JSON not found: {json_path}")
        return

    # 수분 값에 따른 분류
    moisture_category = categorize_moisture(moisture_value)

    # 이미지 파일 경로 설정 (정면 이마)
    image_filename = f"{str(subject_no).zfill(4)}_03_F.jpg"
    image_path = os.path.join(image_folder, str(subject_no).zfill(4), image_filename)

    if os.path.exists(image_path):
        # 이미지 열기
        img = Image.open(image_path)

        # 이마 부분 크롭 및 저장
        if moisture_value is not None and bbox is not None:
            x_min, y_min, x_max, y_max = bbox
            cropped_img = img.crop((x_min, y_min, x_max, y_max))
            destination_path = os.path.join(output_folder, moisture_category, f"{str(subject_no).zfill(4)}_forehead.jpg")
            cropped_img.save(destination_path)
            print(f"Saved forehead image to: {destination_path}")
    else:
        print(f"Image not found: {image_path}")

# 각 subject_no에 대한 이마 이미지를 크롭 및 분류
for index, row in measurements_df.iterrows():
    subject_no = row['subject_no']
    process_forehead(subject_no, row['수분_이마'])
```

```python
import os
import pandas as pd
from PIL import Image
import json

# 이미지와 출력 폴더 경로
image_folder = r'.\image'  # 이미지가 있는 경로
output_folder = r'.\output5'  # 크롭된 이마 이미지가 저장될 경로
bbox_folder = r'.\bbox'  # bbox 정보가 있는 JSON 파일 경로

# 등급별 폴더 생성
categories = ['Very Dry', 'Dry', 'Normal', 'Hydrated', 'Over-hydrated']
for category in categories:
    os.makedirs(os.path.join(output_folder, category), exist_ok=True)

# 수분 값을 기준으로 분류 함수
def categorize_moisture(moisture_value):
    if moisture_value <= 20:
        return 'Very Dry'
    elif 21 <= moisture_value <= 40:
        return 'Dry'
    elif 41 <= moisture_value <= 60:
        return 'Normal'
    elif 61 <= moisture_value <= 80:
        return 'Hydrated'
    else:
        return 'Over-hydrated'
```

```python
# 정면, 좌측, 우측 이마를 크롭하여 저장하는 함수
def process_forehead(subject_no, moisture_value, json_part, image_part, part_name):
    # subject_no를 정수로 변환 후 zfill(4)을 적용하여 4자리 숫자로 변환
    try:
        subject_no_str = str(int(float(subject_no))).zfill(4)  # 정수로 변환 후 4자리로 변환
    except ValueError:
        print(f"Invalid subject_no: {subject_no}")
        return

    # JSON 파일 경로 설정 (예: \bbox\0006\0006_03_L_01.json)
    json_filename = f"{subject_no_str}_03_{json_part}_01.json"
    json_path = os.path.join(bbox_folder, subject_no_str, json_filename)

    # JSON 파일에서 bbox 정보 가져오기
    if os.path.exists(json_path):
        with open(json_path, 'r') as f:
            data = json.load(f)
            bbox = data['images'].get('bbox', None)  # bbox가 존재하지 않으면 None 반환
    else:
        print(f"JSON not found: {json_path}")
        return

    # bbox가 None일 경우 해당 이미지를 건너뜀
    if bbox is None:
        print(f"bbox is None for {json_filename}, skipping...")
        return

    # 수분 값에 따른 분류
    moisture_category = categorize_moisture(moisture_value)

    # 이미지 파일 경로 설정 (정면, 좌측, 우측 이마)
    image_filename = f"{subject_no_str}_03_{image_part}.jpg"
    image_path = os.path.join(image_folder, subject_no_str, image_filename)

    if os.path.exists(image_path):
        # 이미지 열기
        img = Image.open(image_path)

        # 이마 부분 크롭 및 저장
        if moisture_value is not None:
            x_min, y_min, x_max, y_max = bbox
            cropped_img = img.crop((x_min, y_min, x_max, y_max))
            destination_path = os.path.join(output_folder, moisture_category, f"{subject_no_str}_{part_name}.jpg")
            cropped_img.save(destination_path)
            print(f"Saved {part_name} image to: {destination_path}")
    else:
        print(f"Image not found: {image_path}")

# 각 subject_no에 대한 정면, 좌, 우 이마 이미지를 크롭 및 분류
for index, row in measurements_df.iterrows():
    subject_no = row['subject_no']
    
    # 정면 이마(F) 처리
    process_forehead(subject_no, row['수분_이마'], 'F', 'F', 'forehead_front')

    # 좌측 이마(L) 처리
    process_forehead(subject_no, row['수분_이마'], 'L', 'L', 'forehead_left')

    # 우측 이마(R) 처리
    process_forehead(subject_no, row['수분_이마'], 'R', 'R', 'forehead_right')
```

## 평균 & 차이값들

```python
import pandas as pd

# 수분 상태에 따른 점수 매핑
score_mapping = {
    'Very Dry': 1,
    'Dry': 2,
    'Normal': 3,
    'Hydrated': 4,
    'Over-hydrated': 1
}

# 수분 값을 기준으로 분류 함수
def categorize_moisture(moisture_value):
    if moisture_value <= 20:
        return 'Very Dry'
    elif 21 <= moisture_value <= 40:
        return 'Dry'
    elif 41 <= moisture_value <= 60:
        return 'Normal'
    elif 61 <= moisture_value <= 80:
        return 'Hydrated'
    else:
        return 'Over-hydrated'

# 각 subject_no에 대한 이마, 왼쪽 볼, 오른쪽 볼 수분 상태에 따른 점수 계산 함수
def calculate_scores(measurements_df):
    subject_scores = []
    
    for index, row in measurements_df.iterrows():
        subject_no = row['subject_no']
        
        # 이마, 왼쪽 볼, 오른쪽 볼 수분 상태 가져오기
        forehead_moisture = row['수분_이마']
        left_cheek_moisture = row['수분_왼쪽볼']
        right_cheek_moisture = row['수분_오른쪽볼']
        
        # 각 부위별로 점수 계산 (수분 상태를 분류하고 점수 부여)
        forehead_category = categorize_moisture(forehead_moisture)
        left_cheek_category = categorize_moisture(left_cheek_moisture)
        right_cheek_category = categorize_moisture(right_cheek_moisture)
        
        # 부위별 점수 계산
        forehead_score = score_mapping[forehead_category]
        left_cheek_score = score_mapping[left_cheek_category]
        right_cheek_score = score_mapping[right_cheek_category]
        
        # 평균 점수 계산
        average_score = (forehead_score + left_cheek_score + right_cheek_score) / 3
        
        # 각 subject_no의 부위별 점수와 평균 점수를 저장
        subject_scores.append({
            'subject_no': subject_no,
            'forehead_score': forehead_score,
            'left_cheek_score': left_cheek_score,
            'right_cheek_score': right_cheek_score,
            'average_score': average_score
        })
    
    return pd.DataFrame(subject_scores)


# 각 subject_no에 대한 점수 계산 및 평균 계산
subject_scores_df = calculate_scores(measurements_df)

# 결과 출력
print(subject_scores_df[['subject_no', 'forehead_score', 'left_cheek_score', 'right_cheek_score', 'average_score']])
```

```python
import pandas as pd

# 수분 상태에 따른 점수 매핑
score_mapping = {
    'Very Dry': 1,
    'Dry': 2,
    'Normal': 3,
    'Hydrated': 4,
    'Over-hydrated': 1
}

# 수분 값을 기준으로 분류 함수
def categorize_moisture(moisture_value):
    if moisture_value <= 20:
        return 'Very Dry'
    elif 21 <= moisture_value <= 40:
        return 'Dry'
    elif 41 <= moisture_value <= 60:
        return 'Normal'
    elif 61 <= moisture_value <= 80:
        return 'Hydrated'
    else:
        return 'Over-hydrated'

# 각 subject_no에 대한 이마, 왼쪽 볼, 오른쪽 볼 수분 상태에 따른 점수 계산 함수
def calculate_scores(measurements_df):
    subject_scores = []
    
    for index, row in measurements_df.iterrows():
        subject_no = row['subject_no']
        
        # 이마, 왼쪽 볼, 오른쪽 볼 수분 상태 가져오기
        forehead_moisture = row['수분_이마']
        left_cheek_moisture = row['수분_왼쪽볼']
        right_cheek_moisture = row['수분_오른쪽볼']
        
        # 각 부위별로 점수 계산 (수분 상태를 분류하고 점수 부여)
        forehead_category = categorize_moisture(forehead_moisture)
        left_cheek_category = categorize_moisture(left_cheek_moisture)
        right_cheek_category = categorize_moisture(right_cheek_moisture)
        
        # 부위별 점수 계산
        forehead_score = score_mapping[forehead_category]
        left_cheek_score = score_mapping[left_cheek_category]
        right_cheek_score = score_mapping[right_cheek_category]
        
        # 각 subject_no의 부위별 점수 저장
        subject_scores.append({
            'subject_no': subject_no,
            'forehead_score': forehead_score,
            'left_cheek_score': left_cheek_score,
            'right_cheek_score': right_cheek_score
        })
    
    return pd.DataFrame(subject_scores)

# 점수 차이가 2점 이상인 경우 필터링하는 함수
def filter_large_diff(df):
    filtered_df = df[
        (abs(df['forehead_score'] - df['left_cheek_score']) >= 3) |  # 이마 - 왼쪽 볼
        (abs(df['forehead_score'] - df['right_cheek_score']) >= 3) |  # 이마 - 오른쪽 볼
        (abs(df['left_cheek_score'] - df['right_cheek_score']) >= 3)  # 왼쪽 볼 - 오른쪽 볼
    ]
    return filtered_df

# 각 subject_no에 대한 점수 계산 및 데이터프레임 생성
subject_scores_df = calculate_scores(measurements_df)

# 점수 차이가 2점 이상인 경우 필터링
filtered_df = filter_large_diff(subject_scores_df)

# 결과 출력
print(filtered_df[['subject_no', 'forehead_score', 'left_cheek_score', 'right_cheek_score']])
```
