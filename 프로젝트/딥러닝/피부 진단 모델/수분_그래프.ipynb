```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 데이터 로드
file_path = '/content/measurement_data.csv'
data = pd.read_csv(file_path)

# 범위 정의 (5등분, 범위 16씩)
bins = [20, 36, 52, 68, 84, 100]
labels = ['20-36', '36-52', '52-68', '68-84', '84-100']

# 각 부위의 수분 데이터를 범위로 분류
data['이마_range'] = pd.cut(data['수분_이마'], bins=bins, labels=labels, include_lowest=True)
data['왼쪽볼_range'] = pd.cut(data['수분_왼쪽볼'], bins=bins, labels=labels, include_lowest=True)
data['오른쪽볼_range'] = pd.cut(data['수분_오른쪽볼'], bins=bins, labels=labels, include_lowest=True)

# 범위별 개수 세기
count_이마 = data['이마_range'].value_counts().sort_index()
count_왼쪽볼 = data['왼쪽볼_range'].value_counts().sort_index()
count_오른쪽볼 = data['오른쪽볼_range'].value_counts().sort_index()

# 파란색 조정 (84-100 구간은 어두운 파란색으로)
custom_blue_colors = ['#ADD8E6', '#87CEEB', '#4682B4', '#1E90FF', '#36454F']

# 시각화: 막대 그래프 (가로 방향)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 이마 수분 분포 (가로막대)
sns.barplot(y=count_이마.index, x=count_이마.values, ax=axes[0], palette=custom_blue_colors)
axes[0].set_title('이마 수분 분포')
axes[0].set_xlabel('개수')
axes[0].set_ylabel('수분 범위')

# 왼쪽 볼 수분 분포 (가로막대)
sns.barplot(y=count_왼쪽볼.index, x=count_왼쪽볼.values, ax=axes[1], palette=custom_blue_colors)
axes[1].set_title('왼쪽 볼 수분 분포')
axes[1].set_xlabel('개수')
axes[1].set_ylabel('수분 범위')

# 오른쪽 볼 수분 분포 (가로막대)
sns.barplot(y=count_오른쪽볼.index, x=count_오른쪽볼.values, ax=axes[2], palette=custom_blue_colors)
axes[2].set_title('오른쪽 볼 수분 분포')
axes[2].set_xlabel('개수')
axes[2].set_ylabel('수분 범위')

plt.tight_layout()
plt.show()
```
