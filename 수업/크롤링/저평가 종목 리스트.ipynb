```python
#KOSDAQ
sum = []

for n in tqdm(range(1, 47)):

    dic = {"menu" : 'market_sum',
           "returnUrl" : f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={n}",
           "fieldIds" : fields}

    url = requests.post("https://finance.naver.com/sise/field_submit.naver", data = dic)
    html = BeautifulSoup(url.text)

    table = html.find("table", class_ = 'type_2')
    table = pd.read_html(str(table))[0]
    sum.append(table)
    time.sleep(2)
```

```python
data2 = pd.concat(sum)

data2['종류'] = 'KOSDAQ'
```
